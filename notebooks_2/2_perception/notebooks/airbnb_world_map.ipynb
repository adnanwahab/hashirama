```python
df_tmp = pd.read_csv(CSV_PATH)

df_tmp


```

```python
cities = df_tmp['AccentCity'].tolist()
lat = df_tmp['Latitude'].tolist()
lng = df_tmp['Latitude'].tolist()
```

```python
import json

unique_cities = set()

def uniq(i):
    key = str(lat[i]) + str(lng[i])
    if key not in unique_cities:
        unique_cities.add(key)
        return True
    else: return False
        
all_cities = {city:[lat[i], lng[i]] for i, city in enumerate(cities) if uniq(i) }
```

```python
len(all_cities)

json.dump(all_cities, open('../data/all_city_names.json', 'w+'), indent=4)
```
