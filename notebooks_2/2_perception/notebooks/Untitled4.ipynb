```python
#Belize--Belize--Belize.osm
fp = '../osm/Mid-North-Coast--New-South-Wales--Australia.osm'





from lxml import etree

# Load the OSM file
with open(fp, "rb") as f:
    tree = etree.parse(f)

# Find all nodes with an amenity tag
amenity_nodes = tree.xpath("//node[tag/@k='amenity']")

# Print the results or process them further
for node in amenity_nodes:
    node_id = node.get("id")
    amenity_type = node.xpath("tag[@k='amenity']/@v")[0]
    print(f"Node ID: {node_id}, Amenity Type: {amenity_type}")
```
